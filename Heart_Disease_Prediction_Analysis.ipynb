***Introduction to the Heart Disease Dataset***

The Heart Disease dataset is a popular dataset from the UCI Machine Learning Repository. It contains information about individuals, including various medical attributes, and indicates whether they have heart disease or not.

```python
import requests

# Direct URL for the Cleveland dataset (processed)
dataset_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data"

# Send a request to download the dataset
response = requests.get(dataset_url)

# Check if the response is successful
if response.status_code == 200:
    # Save the dataset to a local file
    with open("heart_disease_data.csv", "wb") as file:
        file.write(response.content)
    print("Dataset downloaded and saved as heart_disease_data.csv")
else:
    print(f"Failed to download dataset. Status code: {response.status_code}")
```

****Correlation Matrix****

A correlation matrix is a table that displays the correlation coefficients between pairs of variables in a dataset. It helps to understand how different variables are related to each other. Each cell in the matrix represents the correlation between two variables.

**Inferential Statistics**

**Hypothesis Testing**: Let's test whether the average maximum heart rate achieved thalach differs significantly between patients with and without heart disease.

**Null Hypothesis (H)**: There is no difference in average thalach between patients with and without heart disease.

**Alternative Hypothesis (H)**: There is a significant difference in average thalach between the two groups.

```python
# Separate the data into two groups
from scipy import stats  # Importing the scipy.stats module
thalach_disease = df[df['num'] == 1]['thalach']
thalach_no_disease = df[df['num'] == 0]['thalach']

# Perform an independent t-test
t_stat, p_value = stats.ttest_ind(thalach_disease, thalach_no_disease)

print("T-Statistic:",t_stat)
print("P-Value:",p_value)
```

If the p-value is less than 0.05, we reject the null hypothesis therefore, There is a significant difference in average thalach between the two groups.
